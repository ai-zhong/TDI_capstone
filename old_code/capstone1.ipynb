```python
# this notebook merged patent data and company data for ticker matching
import dill
import pandas as pd
import numpy as np
```

```python
company_dict = dill.load(open('company_dict.pkd', 'rb'))
```

```python
# company_dict: key is name from patent docs; value is name from stock market.
patent_data = dill.load(open('merge_3.pkd', 'rb'))
```

```python
company_dict_s = pd.Series(company_dict)
company_dict_df = company_dict_s.reset_index()
company_dict_df.columns = ['ee_name', 'standard_name']
```

```python
merge_5 = patent_data.merge(company_dict_df, on='ee_name', how='left')
```

```python
ee_match = merge_5[merge_5['standard_name'].map(str) != 'nan']
```

```python
# dill.dump(or_match, open('assignor_matched.pkd', 'wb'))
companylist = pd.read_csv('companylist.csv',index_col=0)
companyinfo_dict = companylist.set_index('Name').T.to_dict()
```

```python
companylist.columns = ['ticker', 'ee_name', 'sector', 'industry']
```

```python
patent_company_ee = ee_match.merge(companylist, on='ee_name', how='inner')
```
